```python
import numpy as np
lst1=[]
lst2=[]
print("For lst1:")
for _ in range(5):
    lst1.append(eval(input()))
print("For lst2:")
for _ in range(5):
    lst2.append(eval(input()))
num_arr1=np.array(lst1)
num_arr2=np.array(lst2)
print(num_arr1+num_arr2) #concatination
print(np.sort(num_arr1))  #num_arr1 sorted
print(np.sort(num_arr2))  #num_arr2 sorted
print(np.sort(num_arr1+num_arr2))  #num_arr1 and num_arr2 concatenated & sorted
```

```python
import numpy as np
print("arr1:")
arr1=np.array([input() for _ in range(4)])
print("arr2:")
arr2=np.array([input() for _ in range(4)])
print()
print(np.vstack((arr1,arr2)))  #vertical stack
print()
print(np.hstack((arr1,arr2)))  #horizontal stack
```
