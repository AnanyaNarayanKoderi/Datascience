```python
import math
position={"x":0,"y":0}
for _ in range(int(input())):
    moves=input().split()
    if moves[0]=="UP":
        position["y"]+=int(moves[1])
    elif moves[0]=="DOWN":
        position["y"]-=int(moves[1])
    elif moves[0]=="LEFT":
        position["x"]-=int(moves[1])
    else:
        position["x"]+=int(moves[1])
print(round(math.sqrt(position["x"]**2 + position["y"]**2)))
```
